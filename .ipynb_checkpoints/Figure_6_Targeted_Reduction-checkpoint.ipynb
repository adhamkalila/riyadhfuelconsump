```python
# Build Gid_2_edge dict
gid_2_edge = {}
for n1,n2 in streets_1.edges():

    gid_2_edge[streets_1[n1][n2]['gid']] = (n1, n2)
```

```python
j = pickle.load(open("data/mx_Riyadh_streets_TAZ_ODs_0_1.pkl", "rb"))
g, od = j.to_igraph()
```

```python
df100_sorted = df100[df100['flow']/4>1][['flow','o_con','d_con', 'path']].sort_values(by=['flow'], ascending=False)
```

```python
paths2gid_map = { e.index : e['gid'] for e in g.es}
```

```python
# The proportion of cars in Riyadh Fleet
Riyadh_fleet_prop = [0.0016624,
0.0003830,
0.0227833,
0.0440881,
0.0487076,
0.0481790,
0.0419814,
0.1098564,
0.0383885,
0.3886037,
0.1585104,
0.0416826,
0.0549435,
0.0002298]

probs = np.array(Riyadh_fleet_prop)
probs /= probs.sum()
```

```python
streets_times = pickle.load(open('data/streets_withtimes.pkl', 'rb'))
```

```python
matched_Edges_1 = []
for e1,e2 in streets_1.edges():
    time1 = streets_1[e1][e2]['FC_bybin_time1']
    if len(time1) > 0: matched_Edges_1.append(streets_1[e1][e2]['gid'])
```

```python
Trips_time1_100[1]
```

```python
mat_edg1=defaultdict(int)
for k in matched_Edges_1:
    mat_edg1[k]=1
```

```python
#  random trip removal based on xtreme increments
No_trips = len(Trips_time1_100)
Random_rank_all_trips_time1_100 = np.random.choice(No_trips, No_trips, replace=False)
```

```python
missing_edges = []

for trip, details in Trips_time1_100.iteritems():
    try:
        details['tripLen_km'] == 0
    except:
        missing_edges.append(trip)
missing_edges.append(745915)
missing_edges.append(373198)
missing_edges.append(744126)
missing_edges.append(377711)
```

```python
limit = 0
Scatter_tt ={}
missing_triptime = []
for trip in Random_rank_all_trips_time1_100:
    if trip in missing_edges: continue
    details = Trips_time1_100[trip]
    o = Trips_time1_100[trip]['o_con']
    d = Trips_time1_100[trip]['d_con']
    limit += 1
    SP_time_m = 0 #[min]
    for gid in details['path']:
        e1,e2 = gid_2_edge[gid]
        if mat_edg1[gid]!=1: continue # skips edges without fuel for all times. For comparison with SP fuel cons
        SP_time_m += streets_times[e1][e2]['t_total'][0]/60 #[min]

    details['Taxi_triptime1_m'] = SP_time_m # [min]
    try: Trips_time1_100[trip]['triptime1_m']
    except: 
        missing_triptime.append(trip)
        continue
    Scatter_tt[trip]={'OD' : (o,d),'CS_assign_time_m':Trips_time1_100[trip]['triptime1_m'], 'Taxi_time_m': SP_time_m}
#     if limit == 100000: break
```

```python
SP_time_m=[]
done = []
CS_assign_time_m=[]
```

```python
# make a list to plot. takes about 1.5 hours
ttt= time.time()

for t, times in Scatter_tt.iteritems():
    if times['OD'] in done: continue
    if times['CS_assign_time_m'] != 0:
        SP_time_m.append(times['Taxi_time_m'])
        CS_assign_time_m.append(times['CS_assign_time_m'])
    done.append(times['OD'])
print 'time elapsed', round(time.time() - ttt), 'sec'
```

```python
# Load segments mathced by edge
segments_byedge = pickle.load(open('data/matchedSegmentsbyEdge_time1_8_9AM_weekdays_github.pkl', 'rb'))
```

```python
# start with trips no 1, 50000, 500000 in Trips pl
plots = [660000, 530000,780000]
SP_plots = {}
for trip in plots:
    path1 = Trips_time1_100[trip]['path']
    lenp = len(path1)
    skipped =0
    times = [0] 
    comptime = 0
    speeds = [0]
    for gid in path1:
        if not gid in segments_byedge.keys(): 
            skipped += 1
            continue
        for tripid, segs in segments_byedge[gid].iteritems():
            for cell in segs[0]:
                comptime += cell[3]  #[sec]
                if cell[5] > 160: continue
                times.append(comptime)
                speeds.append(cell[5]) # [km/h]
            break
    SP_plots[trip] = {'speeds':speeds, 'times':times}
            
        
```

```python

Speedprof_kmh1 = SP_plots[plots[0]]['speeds']
Speedprof_sec1 = SP_plots[plots[0]]['times']
Speedprof_sec1_new = np.linspace(min(Speedprof_sec1),max(Speedprof_sec1),400) #600 represents number of points to make between T.min and T.max
Speedprof_kmh1_new = spline(Speedprof_sec1,Speedprof_kmh1,Speedprof_sec1_new)


Speedprof_kmh2 = SP_plots[plots[1]]['speeds']
Speedprof_sec2 = SP_plots[plots[1]]['times']
Speedprof_sec2_new = np.linspace(min(Speedprof_sec2),max(Speedprof_sec2),600) #600 represents number of points to make between T.min and T.max
Speedprof_kmh2_new = spline(Speedprof_sec2,Speedprof_kmh2,Speedprof_sec2_new)

Speedprof_kmh3 = SP_plots[plots[2]]['speeds']
Speedprof_sec3 = SP_plots[plots[2]]['times']
Speedprof_sec3_new = np.linspace(min(Speedprof_sec3),max(Speedprof_sec3),600) #600 represents number of points to make between T.min and T.max
Speedprof_kmh3_new = spline(Speedprof_sec3,Speedprof_kmh3,Speedprof_sec3_new)

```

```python
#removing missing trips
for trip in sorted(missing_edges, reverse=True):
    del Trips_time1_100[trip]
```

```python
# clean FC_alltrips
FC_alltrips = [details['FC_time1'] for trip, details in Trips_time1_100.iteritems() if details['tripLen_km']>0 and details['FC_time1'] >0 and details['FC_time1'] < 15]
# creating fuel economy for hist
CS_FE_alltrips = [details['CS_FC_tripd']**(-1) for trip, details in Trips_time1_100.iteritems() if details['tripLen_km']>0 and details['FC_time1'] >0 ]
FE_alltrips = [details['FC_tripd']**(-1) for trip, details in Trips_time1_100.iteritems() if details['tripLen_km']>0 and details['FC_time1'] >0]
```

```python
# Extreme number of increments 

trips_removal_incr_xtreme = np.arange(0,800000,10000)
```

```python
#  random trip removal based on xtreme increments
No_trips = len(Trips_time1_100)
Random_rank_all_trips_time1_100 = np.random.choice(No_trips, No_trips, replace=False)

FC_reduction_time1_rand_xtreme = [0 for a in range(len(trips_removal_incr_xtreme))]
CS_FC_reduction_time1_rand_xtreme = [0 for a in range(len(trips_removal_incr_xtreme))]
trips_removal_incr_xtreme[0]=1

curr_red = 0
alpha = 0
FC_reduce= 0
CS_FC_reduce = 0
for trip in Random_rank_all_trips_time1_100:
    if trip in missing_edges: continue
    try: Trips_time1_100[trip]['FC_time1']
    except: continue 
    FC_reduce += Trips_time1_100[trip]['FC_time1']
    CS_FC_reduce += Trips_time1_100[trip]['CS_FC_time1']
    curr_red += 1
    while curr_red == trips_removal_incr_xtreme[alpha]: 
        FC_reduction_time1_rand_xtreme[alpha] = FC_reduce
        CS_FC_reduction_time1_rand_xtreme[alpha] = CS_FC_reduce
        if alpha == len(trips_removal_incr_xtreme)-1: break
        alpha += 1
```

```python
# Rank based on FC per distance
trip_no_rank_FC_per_m = []
for trip in sorted(Trips_time1_100, key = lambda k: Trips_time1_100[k]['FC_tripd'], reverse=True): # produces a list sorted by values in that attribute
    trip_no_rank_FC_per_m.append(trip)
    
       
FC_reduction_time1_targ_FC_tripd_xtreme = [0 for a in range(len(trips_removal_incr_xtreme))]

curr_red = 0
alpha = 0
FC_reduce= 0
for trip in trip_no_rank_FC_per_m:
    if trip in missing_edges: continue
    FC_reduce += Trips_time1_100[trip]['FC_time1']
    curr_red += 1
    while curr_red == trips_removal_incr_xtreme[alpha]: 
        FC_reduction_time1_targ_FC_tripd_xtreme[alpha] = FC_reduce
        if alpha == len(trips_removal_incr_xtreme)-1: break
        alpha += 1
```

```python
# Rank based on constant speed CS (Streetsmart) FC per trip dist

trip_no_rank_CS_FC_per_m = []
for trip in sorted(Trips_time1_100, key = lambda k: Trips_time1_100[k]['CS_FC_tripd'], reverse=True): # produces a list sorted by values in that attribute
    trip_no_rank_CS_FC_per_m.append(trip)
    
    
FC_reduction_time1_targ_CS_FC_tripd_xtreme = [0 for a in range(len(trips_removal_incr_xtreme))]

curr_red = 0
alpha = 0
FC_reduce= 0
for trip in trip_no_rank_CS_FC_per_m:
    if trip in missing_edges: continue
    FC_reduce += Trips_time1_100[trip]['CS_FC_time1']
    curr_red += 1
    while curr_red == trips_removal_incr_xtreme[alpha]: 
        FC_reduction_time1_targ_CS_FC_tripd_xtreme[alpha] = FC_reduce
        if alpha == len(trips_removal_incr_xtreme)-1: break
        alpha += 1

```

```python
# Total Fuel consumption 
FC_total_time1_100 = 0
CS_FC_total_time1_100 = 0
for trip, details in Trips_time1_100.iteritems():
    CS_FC_total_time1_100 += details['CS_FC_time1']
    FC_total_time1_100 += details['FC_time1']
```

```python
# Ratio of Fuel with reduction over fuel original 
FC_reduction_time1_rand_xtreme_ratio = [1 - (FC_reduction_time1_rand_xtreme[i]/FC_total_time1_100) for i in range(len(trips_removal_incr_xtreme))]
FC_reduction_time1_targ_FC_tripd_xtreme_ratio = [1 - (FC_reduction_time1_targ_FC_tripd_xtreme[i]/FC_total_time1_100) for i in range(len(trips_removal_incr_xtreme))]
FC_reduction_time1_targ_CS_FC_tripd_xtreme_ratio = [1 - (FC_reduction_time1_targ_CS_FC_tripd_xtreme[i]/CS_FC_total_time1_100) for i in range(len(trips_removal_incr_xtreme))]
```

```python

# constant speed lines for same trips
x1 = np.arange(Trips_time1_100[plots[0]]['triptime1_m'])
y1 = [Trips_time1_100[plots[0]]['tripLen_km']/(Trips_time1_100[plots[0]]['triptime1_m']/60) for i in range(len(x1))]
# y1 = [np.mean(Speedprof_kmh1) for i in range(len(x1))]

x2 = np.arange(Trips_time1_100[plots[1]]['triptime1_m'])
y2 = [Trips_time1_100[plots[1]]['tripLen_km']/(Trips_time1_100[plots[1]]['triptime1_m']/60) for i in range(len(x2))]

x3 = np.arange(Trips_time1_100[plots[2]]['triptime1_m'])
y3 = [Trips_time1_100[plots[2]]['tripLen_km']/(Trips_time1_100[plots[2]]['triptime1_m']/60) for i in range(len(x3))]
```

```python
CS_alltrips = [details['CS_FC_time1'] for trip, details in Trips_time1_100.iteritems() if details['tripLen_km']>0 and details['FC_time1'] >0]
```

```python
# calculating residuals for non normalized version of plot 6d
FC_residual_rand_xtreme = [FC_total_time1_100 - FC_reduction_time1_rand_xtreme[i] for i in range(len(trips_removal_incr_xtreme))]
CS_FC_residual_targ_xtreme = [CS_FC_total_time1_100 - FC_reduction_time1_targ_CS_FC_tripd_xtreme[i] for i in range(len(trips_removal_incr_xtreme))]

FC_residual_targ_xtreme = [FC_total_time1_100 - FC_reduction_time1_targ_FC_tripd_xtreme[i] for i in range(len(trips_removal_incr_xtreme))]

```
